# City Data

```python
data_city= pd.read_csv('Historical weather data/city_attributes.csv')
data_city.head()
type(data_city)
```

# Humidity_Data

```python
data2=pd.read_csv('Historical weather data/humidity.csv')
data2=data2.set_index('datetime')
data2
```

```python
pivot=pd.DataFrame(data2.unstack(level=0))
data2_clean=pivot.reset_index().rename(columns={'level_0':'City', 0:'Humidity %'})
data2_clean.head()
```

```python
data_humidity=data2_clean.dropna(axis=0,how='any')
data_humidity.head()
```

# Pressure Data

```python
data3=pd.read_csv('Historical weather data/pressure.csv')
data3=data3.set_index('datetime')
data3
```

```python
pivot2=pd.DataFrame(data3.unstack(level=0))
data3_clean=pivot.reset_index().rename(columns={'level_0':'City', 0:'Pressure'})
data_pressure=data3_clean.dropna(axis=0,how='any')

data_pressure.head()
```

# temperature

```python
data4=pd.read_csv('Historical weather data/temperature.csv')
data4=data4.set_index('datetime')

pivot3=pd.DataFrame(data4.unstack(level=0))
data4_clean=pivot3.reset_index().rename(columns={'level_0':'City', 0:'Temperature in Kelvin'})

data_temp=data4_clean.dropna(axis=0,how='any')

data_temp.head()
```

# weather description

```python
data5=pd.read_csv('Historical weather data/weather_description.csv')
data5=data5.set_index('datetime')


pivot4=pd.DataFrame(data5.unstack(level=0))
data5_clean=pivot4.reset_index().rename(columns={'level_0':'City', 0:'weather'})

data_weather=data5_clean.dropna(axis=0,how='any')

data_weather.head()
```

# wind speed

```python
data6=pd.read_csv('Historical weather data/wind_speed.csv')
data6=data6.set_index('datetime')


pivot5=pd.DataFrame(data6.unstack(level=0))
data6_clean=pivot5.reset_index().rename(columns={'level_0':'City', 0:'wind speed'})

data_wind_speed=data6_clean.dropna(axis=0,how='any')

data_wind_speed.head()
```

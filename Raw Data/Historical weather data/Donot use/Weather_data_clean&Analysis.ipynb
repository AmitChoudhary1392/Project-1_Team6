```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#read weather_merged csv

input_file= pd.read_csv('weather_merged.csv')
input_file.head()
```

```python
df=input_file.loc[:,['City','Country', 'Latitude','Longitude', 'datetime', 'Temperature in Kelvin']]
df=df.rename(columns={'datetime':'Timestamp', 'Temperature in Kelvin':'Temperature(K)'})
df.head()
```

```python
   
df['Temperature(C)']= round(df.loc[:,'Temperature(K)']- 273.15,2)

df.head()
```

```python
df.count()
```

```python
df['datetimeobj'] =  pd.to_datetime(df['Timestamp'], format='%Y-%m-%d %H:%M:%S.%f')
df.head()
```

```python
df['Hour']= df['datetimeobj'].dt.hour
df['Year']=df['datetimeobj'].dt.year
df['Month']=df['datetimeobj'].dt.month

df=df.loc[:,['datetimeobj','Year','Month','Hour','Country','City','Latitude','Longitude','Temperature(K)','Temperature(C)']]

df.head()
```

```python
# Weather Data Analysis for NYC

#data_NYC= df.loc[:, ['datetimeobj','Year','Month','Hour','Country','City'=='Los Angeles','Latitude','Longitude','Temperature(K)','Temperature(C)']]
data_NYC=df.loc[df['City']=='New York',:]
data_NYC.head()
```
